```python
# Задание 9
df = pd.read_csv('Homework2.csv', sep=';', parse_dates=['Date'], dayfirst=True)
df1 = pd.read_excel('Homework5.xlsx', parse_dates=['Date'], dayfirst=False)

print(df.head())
print(df1.head())
```

```python
df3 = df.merge(df1, left_on='Date', right_on='Date')
df3.head()
```

```python
df3.to_csv('result5.csv', sep='\t')
```

```python
# Задание 10
df3['Temperature_C'] = (df['Temperature'] - 32) * (5/9)
df3.head()
```

```python
df3.to_csv('result6.csv')
```

```python
# Задание 11
fig, ax = plt.subplots(figsize = (12, 6))

plt.scatter(df3['Sales'], df3['Temperature_C'])

ax.set_title('Зависимость количества продаж от температуры')
ax.set_xlabel('Количество продаж')
ax.set_ylabel('Температура (Цельсия)')

plt.show()
```

```python
# Количество продаж очень сильно зависят от температуры на улице, ибо чем ниже температура, тем меньше 
# людей приходят в магазин

# Я думаю что тут приведены продажи легких курток на лето, т.к чем выше температура, тем больше продаж, 
# а температура в некоторых местах превышает 30 градусоув цельсия
```
